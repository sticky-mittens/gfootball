```python
from matplotlib import pyplot as plt 
import seaborn as sns
import numpy as np
sns.set_theme(style="darkgrid")
plt.rc('font', size=20)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=30)  # fontsize of the figure title

def plot():
    # x = ['Option-Templates (11) @ ', 'Baseline (11) @ ', 'IMPALA (1) @ 500 M', 'DQN (1) @ 500 M']
    x = np.array([1,2,3,4])
    easy_y = np.array([5.79, 1.2, 5.14, 8.16]) # best easy is with output_longRun_easy_0.99.log on ubuntu laptop
    easy_yerr = np.array([0.92, 0.33, 2.88, 1.05])

    medium_y = np.array([3.0, 0.8, -0.36, 2.01])
    medium_yerr = np.array([0.71, 0.33, 0.11, 0.27])

    hard_y = np.array([2.0, 0.26, -0.47, 0.27])
    hard_yerr = np.array([0.18, 0.1, 0.48, 0.56])
    
    fig, ax = plt.subplots(figsize=(9,4))
    w = 0.2
    ax.bar(x-w, easy_y, width=w*5/6, yerr=easy_yerr, color='b')
    ax.bar(x, medium_y, width=w*5/6, yerr=medium_yerr, color='y')
    ax.bar(x+w, hard_y, width=w*5/6, yerr=hard_yerr, color='r')
    # plt.ylim([-2, 10])
    # ax.set_title('')
    ax.set_ylabel('avg goal difference')
    # ax.set_xlabel('My xlabel')
    # ax.tick_params(axis='x', rotation=60)
    plt.savefig(f'./result_plots/gfootball_bar_plot.png')

plot()
```
